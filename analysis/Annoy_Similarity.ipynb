```python
from annoy_similarity import AnnoySimilarity
import pandas as pd
from IPython.display import display, HTML
import time
```

```python
large_all = '/Users/amandeep/Github/maa-analysis/MAA_Datasets/v3.2.0/text_embeddings_all_large_v3.tsv.gz'
large_2000 = '/Users/amandeep/Github/maa-analysis/MAA_Datasets/v3.2.0/embedding_files/2000.tsv'
wiki_labels = '/Users/amandeep/Github/maa-analysis/MAA_Datasets/v3.2.0/qnodes-properties-labels-for-V3.2.0_KB.tsv'
```

```python
s = time.time()
ai = AnnoySimilarity(large_all, wiki_labels, annoy_index_path='annoy_index_dot_all.ann', build_new_index=True, metric='dot')
print(time.time() - s)
```

```python
queries = ['Q37828', 'Q48989064', 'Q271997', 'Q48989064', 'Q159683', 'Q127956', 'Q60']
# queries = ['Q97', 'Q99']
df_list = []
for q in queries:
    df_list.append(pd.DataFrame(ai.nearest_neighbors(q, k=10)))
df = pd.concat(df_list)
display(HTML(df.to_html()))
```

```python
df.to_csv('annoy_dot_results.csv', index=False)
```
